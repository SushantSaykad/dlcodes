```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to compute the value of the function y = (x+3)²
def compute_function(x):
    return (x + 3) ** 2
```

```python
# Function to compute the derivative of the function y = (x+3)²
def compute_derivative(x):
    return 2 * (x + 3)
```

```python
# Function to perform Gradient Descent to find the local minima
def gradient_descent(x, learning_rate, iterations):
    for i in range(iterations):
        gradient = compute_derivative(x)
        x = x - learning_rate * gradient
    return x
```

```python
# Function to plot the function and the gradient descent progress
def plot_function_and_gradient_descent(x_vals, y_vals, x_minima):
    x = np.linspace(-10, 10, 400)
    y = compute_function(x)

    plt.plot(x, y, label='Function: y=(x+3)²')
    plt.scatter(x_vals, y_vals, c='r', label='Gradient Descent')
    plt.scatter(x_minima, compute_function(x_minima), c='g', marker='x', label='Local Minima')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Gradient Descent to Find Local Minima')
    plt.show()
```

```python
# Main code
starting_point = 2.0
learning_rate = 0.1
iterations = 100

x_vals = []
y_vals = []

x = starting_point
for i in range(iterations):
    y = compute_function(x)
    x_vals.append(x)
    y_vals.append(y)
    gradient = compute_derivative(x)
    x = x - learning_rate * gradient

local_minima = x
min_value = compute_function(local_minima)

print('Local Minima:', local_minima)
print('Min Value:', min_value)

plot_function_and_gradient_descent(x_vals, y_vals, local_minima)
```
